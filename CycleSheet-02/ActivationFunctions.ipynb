Q) List all Activation Functions

```python
## 1 Binary Step
def binary_step(x):
    if x<0:
        return 0
    else:
        return 1

import matplotlib.pyplot as plt

## define input data
x = [i for i in range(-10,13)]
y = [binary_step(i) for i in x]
plt.plot(x,y)
plt.title('Binary Step Activation Function')
plt.show()
```

```python
## Linear Activation Function
def linear_function(x):
    return 4*x

## define input data
x = [i for i in range(-10,13)]
y = [linear_function(i) for i in x]
plt.plot(x,y)
plt.title('Linear Activation Function')
plt.show()
```

```python
## sigmoid activation function
import numpy as np

def sigmoid(x):
    z = (1/(1+np.exp(-x)))
    return z

## define input data
x = [i for i in range(-10,13)]
y = [sigmoid(i) for i in x]
plt.plot(x,y)
plt.title('Sigmoid Activation Function')
plt.show()
```

```python
## tanh activation function
def tanh(x):
    z = (2/(1 + np.exp(-2*x))) -1
    return z

## define input data
x = [i for i in range(-10,13)]
y = [tanh(i) for i in x]
plt.plot(x,y)
plt.title('Tanh Activation Function')
plt.show()
```

```python
## relu activation function
def relu(x):
    return max(0.0,x)

## define input data
x = [i for i in range(-10,13)]
y = [relu(i) for i in x]
plt.plot(x,y)
plt.title('Relu Activation Function')
plt.show()
```

```python
## leaky Relu
def leaky_relu(x):
    if x<0:
        return 0.01*x
    else:
        return x

## define input data
x = [i for i in range(-10,13)]
y = [leaky_relu(i) for i in x]
plt.plot(x,y)
plt.title('Leaky Relu Activation Function')
plt.show()
```

```python
## parameterized Relu
def para_relu(x):
    if x>=0:
        return x
    else:
        return 3*x

## define input data
x = [i for i in range(-10,13)]
y = [para_relu(i) for i in x]
plt.plot(x,y)
plt.title('Parameterized Relu Activation Function')
plt.show()
```

```python
## exponential linear unit
def expo(x, a):
    if x<0:
        return a*(np.exp(x)-1)
    else:
        return x

## define input data
x = [i for i in range(-10,13)]
y = [expo(i, 0.5) for i in x]
plt.plot(x,y)
plt.title('Exponential Linear Activation Function')
plt.show()
```

```python
## swish activation function
def swish(x):
    return x/(1-np.exp(-x))

## define input data
x = [i for i in range(-10,13)]
y = [swish(i) for i in x]
plt.plot(x,y)
plt.title('Swish Activation Function')
plt.show()
```
